```python
import cv2
import matplotlib.pyplot as plt
from deepface import DeepFace

# Load the pre-trained model
model_name = "Facenet"
model = DeepFace.build_model(model_name)

def verify(image1, image2,model):
    """
    Verify whether two images are the same person or not using DeepFace's verification model.
    
    Parameters:
    image1 (str or numpy.ndarray): file path or image array of the first image.
    image2 (str or numpy.ndarray): file path or image array of the second image.
    
    Returns:
    None
    """
    try:
        # Read input images if they are file paths
        if isinstance(image1, str):
            img1 = cv2.imread(image1)
        else:
            img1 = image1

        if isinstance(image2, str):
            img2 = cv2.imread(image2)
        else:
            img2 = image2

        # Display input images
        plt.subplot(1, 2, 1)
        plt.imshow(img1[:, :, ::-1])
        plt.title('Image 1')
        plt.axis('off')

        plt.subplot(1, 2, 2)
        plt.imshow(img2[:, :, ::-1])
        plt.title('Image 2')
        plt.axis('off')

        plt.show()

        # Use DeepFace's verification model to compare the images
        output = DeepFace.verify(img1, img2, model_name=model)

        verification = output['verified']

        if verification:
            print("Images are of the same person.")
        else:
            print("Images are of different people.")

    except Exception as e:
        print(f"Error: {e}")
```


# 2. Face_recognition library
The face_recognition library is built on deep learning techniques and uses only a single training image.

How face_recognition works
Passes a person's picture with their name to the model.
The model takes every picture and after converting them to numerical encoding, stores them in a list with the labels in another list.
When making a prediction, the model again converts the input image to an encoding.
This encoded image is used to find similar encodings based on the distance. The least distant encodings will be the closest match for the person.
Once the closest match is found, the final detected person’s name is produced.

```python
import cv2
import face_recognition

# Load the image as a NumPy array
image = cv2.imread('/content/drive/MyDrive/Colab Notebooks/face recognization/images/111.jpg')

# Convert the image from OpenCV format (BGR) to RGB format
rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Find faces in the image and get their locations
face_locations = face_recognition.face_locations(rgb_image)

# Draw rectangles around the detected faces
for (top, right, bottom, left) in face_locations:
    cv2.rectangle(image, (left, top), (right, bottom), (0, 255, 0), 2)

# Display the image with detected faces
cv2_imshow(image)
cv2.waitKey(0)
```
